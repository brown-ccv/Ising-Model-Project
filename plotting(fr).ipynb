```julia
import Pkg
Pkg.add("Plots")
Pkg.add("DelimitedFiles")

using Plots
using DelimitedFiles

function read_xy_data(filename, delimiter='\t')
    data = readdlm(filename, delimiter)
    x_data = data[:, 1]
    y_data = data[:, 2]
    return x_data, y_data
end
filename = "1sr.txt" # Replace with the name of your text file
filenames = "2nd.txt" # Replace with the name of your text file
x_data, y_data = read_xy_data(filename)
x1_data, y1_data = read_xy_data(filenames)

plot([x_data; x1_data], [y_data; y1_data], xlabel="x-axis", ylabel="y-axis", title="X-Y Data Plot", legend=false)
```
